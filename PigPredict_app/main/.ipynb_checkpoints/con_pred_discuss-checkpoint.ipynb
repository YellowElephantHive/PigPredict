```python
""" load pickle """
con_pred_path = os.path.join(prj_root, 'resource', 'pickled', 'cons_pred.pkl')
concentration_path = os.path.join(prj_root, 'resource', 'pickled', 'decoder_x_val.pkl')
lab_path = os.path.join(prj_root, 'resource', 'pickled', 'decoder_y_val.pkl')
lab_mean_path = os.path.join(prj_root, 'resource', 'pickled', 'lab_mean.pkl')
lab_std_path = os.path.join(prj_root, 'resource', 'pickled', 'lab_std.pkl')
decoder_input_columns_path = os.path.join(prj_root, 'resource', 'pickled', 'decoder_input_columns.pkl')

with open(lab_mean_path, 'rb') as f:
    lab_mean = pickle.load(f)
with open(lab_std_path, 'rb') as f:
    lab_std = pickle.load(f)
with open(con_pred_path, 'rb') as f:
    pred = pickle.load(f)
with open(concentration_path, 'rb') as f:
    concentration = pickle.load(f)
with open(lab_path, 'rb') as f:
    lab = pickle.load(f)
with open(decoder_input_columns_path, 'rb') as f:
    decoder_input_columns = pickle.load(f)
```

```python
""" 參數設定 """
con_dim = len(decoder_input_columns)
clothe_type_count = 3
DATA_FILE_SINGLE = os.path.join(prj_root, 'data', 'hong_make_single_revised_3.csv')
```

```python
""" load data """
df_single = pd.read_csv(DATA_FILE_SINGLE)
df_single = df_single[df_single['abort']!=1]
```

```python
""" 調整預測濃度 """
pred[(pred < 5e-3) & (pred!=0)] = 1e-2
pred[pred>3] = 3
```

```python
target = concentration.values[:len(pred),]
```

```python
diff = pred - target
```

```python
""" 計算平均 mae 和 mse """
# 計算染料數量... 要先把布料數量扣掉
con_not_zero_count = (pred!=0).sum() - (pred[:,int(-1*clothe_type_count):]>0).sum()
mae_all = np.abs(diff).sum()/(con_not_zero_count)
mse_all = (diff ** 2).sum() /(con_not_zero_count)
print(mae_all)
print(mse_all)
```

```python
row_maes = np.abs(diff).sum(axis=1)/((pred!=0).sum(axis=1)-1)
sort_idx = np.argsort(row_maes)
row_maes_sort = row_maes[sort_idx]
pred_sort = pred[sort_idx]
target_sort = target[sort_idx]
diff_sort = diff[sort_idx]
lab_sort = unnormalize_lab(lab.copy().values[sort_idx])
lch_sort = color.lab2lch(lab_sort)
```

畫出 error 和 lch 的關係

```python
def plot_scatter_and_regression(x, y, x_label='', y_label='', bin_size=10):
    batch_x = [x[i:i+bin_size].mean() for i in range(0, len(x), bin_size)]
    batch_x = np.array(batch_x)
     
    batch_y = [y[i:i+bin_size].mean() for i in range(0, len(y), bin_size)]
    batch_y = np.array(batch_y)
    
    r = linregress(batch_x, batch_y)
    
    f, ax = plt.subplots(1,1)

    ax.scatter(batch_x, batch_y)
    ax.plot(batch_x, r.slope * batch_x + r.intercept, '-c')
    
    anchored_text = AnchoredText(f'rvalue:{r.rvalue:.2f}\npvalue:{r.pvalue:.2e}', loc=2)
    ax.add_artist(anchored_text)

    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    
    plt.show()    
```

```python
# log(mae) vs L
plot_scatter_and_regression(np.log(row_maes_sort), lch_sort[:,0], bin_size=5)
```

```python
# log(mae) vs C
plot_scatter_and_regression(np.log(row_maes_sort), lch_sort[:,1], bin_size=1)
```

```python
# log(mae) vs H
plot_scatter_and_regression(np.log(row_maes_sort), lch_sort[:,2], bin_size=5)
```

```python
interest_idx = -12
output_pred = np.array([pred_sort[interest_idx], target_sort[interest_idx]])
lch_sort[interest_idx]
```

```python
output_columns = []
for i in range(1, 5):
    output_columns += [f'染料_{i}', f'濃度_{i}']
output_columns += ['布號']
df_output = pd.DataFrame(columns=output_columns)

clothe_type = decoder_input_columns[np.where(pred_sort[-1]==1)[0][-1]]

for i, con in enumerate(output_pred[:, :con_dim-clothe_type_count]):
    dye_idx = np.where(con > 0)
    dyes = decoder_input_columns[dye_idx]
    dye_cons = con[dye_idx]
    for j in range(len(dye_idx[0])):
        df_output.loc[i, [f'染料_{j+1}', f'濃度_{j+1}']] = [dyes[j].split('_')[1], f'{dye_cons[j]:.4f}']

df_output['布號'] = clothe_type
df_output
```
